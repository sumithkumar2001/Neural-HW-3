```python
#Question 1
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.optimizers import Adam

# Load and preprocess MNIST data
(x_train, _), (x_test, _) = mnist.load_data()
x_train = x_train.astype("float32") / 255.
x_test = x_test.astype("float32") / 255.
x_train = x_train.reshape((len(x_train), 784))
x_test = x_test.reshape((len(x_test), 784))

# Autoencoder builder function
def build_autoencoder(latent_dim):
    input_img = Input(shape=(784,))
    encoded = Dense(latent_dim, activation='relu')(input_img)
    decoded = Dense(784, activation='sigmoid')(encoded)
    autoencoder = Model(input_img, decoded)
    return autoencoder

# Function to train and visualize results for different latent dimensions
def run_experiment(latent_dims):
    for latent_dim in latent_dims:
        print(f"\nTraining Autoencoder with latent_dim = {latent_dim}")
        autoencoder = build_autoencoder(latent_dim)
        autoencoder.compile(optimizer=Adam(), loss='binary_crossentropy')

        autoencoder.fit(x_train, x_train,
                        epochs=10,
                        batch_size=256,
                        shuffle=True,
                        validation_data=(x_test, x_test),
                        verbose=0)

        decoded_imgs = autoencoder.predict(x_test)

        # Plot original vs reconstructed images
        n = 10
        plt.figure(figsize=(20, 4))
        for i in range(n):
            # Original
            ax = plt.subplot(2, n, i + 1)
            plt.imshow(x_test[i].reshape(28, 28), cmap="gray")
            plt.axis("off")

            # Reconstructed
            ax = plt.subplot(2, n, i + 1 + n)
            plt.imshow(decoded_imgs[i].reshape(28, 28), cmap="gray")
            plt.axis("off")
        plt.suptitle(f"Latent Dimension: {latent_dim}")
        plt.show()

# Run with various latent dimensions
run_experiment([16, 32, 64])
```

```python
#Question 2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.optimizers import Adam

# Load and preprocess MNIST data
(x_train, _), (x_test, _) = mnist.load_data()
x_train = x_train.astype("float32") / 255.
x_test = x_test.astype("float32") / 255.
x_train = x_train.reshape((len(x_train), 784))
x_test = x_test.reshape((len(x_test), 784))

# Add Gaussian noise to inputs
def add_noise(images, noise_factor=0.5):
    noisy = images + np.random.normal(loc=0.0, scale=noise_factor, size=images.shape)
    return np.clip(noisy, 0., 1.)

x_train_noisy = add_noise(x_train)
x_test_noisy = add_noise(x_test)

# Autoencoder builder
def build_autoencoder(latent_dim):
    input_img = Input(shape=(784,))
    encoded = Dense(latent_dim, activation='relu')(input_img)
    decoded = Dense(784, activation='sigmoid')(encoded)
    autoencoder = Model(input_img, decoded)
    return autoencoder

# Train and visualize denoising autoencoder
def run_denoising_autoencoder(latent_dim=32):
    print("Training Denoising Autoencoder...")
    autoencoder = build_autoencoder(latent_dim)
    autoencoder.compile(optimizer=Adam(), loss='binary_crossentropy')

    autoencoder.fit(x_train_noisy, x_train,
                    epochs=10,
                    batch_size=256,
                    shuffle=True,
                    validation_data=(x_test_noisy, x_test),
                    verbose=0)

    decoded_imgs = autoencoder.predict(x_test_noisy)

    # Plot noisy vs. reconstructed
    n = 10
    plt.figure(figsize=(20, 6))
    for i in range(n):
        # Noisy input
        ax = plt.subplot(3, n, i + 1)
        plt.imshow(x_test_noisy[i].reshape(28, 28), cmap="gray")
        plt.title("Noisy")
        plt.axis("off")

        # Reconstructed
        ax = plt.subplot(3, n, i + 1 + n)
        plt.imshow(decoded_imgs[i].reshape(28, 28), cmap="gray")
        plt.title("Denoised")
        plt.axis("off")

        # Ground truth
        ax = plt.subplot(3, n, i + 1 + 2 * n)
        plt.imshow(x_test[i].reshape(28, 28), cmap="gray")
        plt.title("Original")
        plt.axis("off")
    plt.suptitle("Denoising Autoencoder Results", fontsize=16)
    plt.show()

# Compare with basic autoencoder
def compare_basic_vs_denoising(latent_dim=32):
    print("Training Basic Autoencoder...")
    basic_autoencoder = build_autoencoder(latent_dim)
    basic_autoencoder.compile(optimizer=Adam(), loss='binary_crossentropy')
    basic_autoencoder.fit(x_train, x_train,
                          epochs=10,
                          batch_size=256,
                          shuffle=True,
                          validation_data=(x_test, x_test),
                          verbose=0)

    basic_decoded = basic_autoencoder.predict(x_test_noisy)

    # Plot comparison
    n = 10
    plt.figure(figsize=(20, 6))
    for i in range(n):
        # Noisy input
        ax = plt.subplot(3, n, i + 1)
        plt.imshow(x_test_noisy[i].reshape(28, 28), cmap="gray")
        plt.title("Noisy")
        plt.axis("off")

        # Basic autoencoder
        ax = plt.subplot(3, n, i + 1 + n)
        plt.imshow(basic_decoded[i].reshape(28, 28), cmap="gray")
        plt.title("Basic Output")
        plt.axis("off")

        # Ground truth
        ax = plt.subplot(3, n, i + 1 + 2 * n)
        plt.imshow(x_test[i].reshape(28, 28), cmap="gray")
        plt.title("Original")
        plt.axis("off")
    plt.suptitle("Basic Autoencoder on Noisy Input", fontsize=16)
    plt.show()

# Run both
run_denoising_autoencoder()
compare_basic_vs_denoising()
```

```python
#Question 3
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding
import requests
import random
import matplotlib.pyplot as plt

# Load a sample text dataset (The Little Prince)
url = "https://www.gutenberg.org/files/60102/60102-0.txt"
text = requests.get(url).text.lower()
text = text[1000:10000]  # Truncate to a smaller chunk for quick training

# Create character-to-index mappings
chars = sorted(list(set(text)))
char2idx = {c: i for i, c in enumerate(chars)}
idx2char = {i: c for i, c in enumerate(chars)}

vocab_size = len(chars)
print(f"Unique characters: {vocab_size}")

# Create sequences for training
seq_length = 40
step = 3
sequences = []
next_chars = []

for i in range(0, len(text) - seq_length, step):
    sequences.append(text[i: i + seq_length])
    next_chars.append(text[i + seq_length])

print(f"Number of sequences: {len(sequences)}")

# Vectorize sequences
X = np.zeros((len(sequences), seq_length, vocab_size), dtype=np.bool_)
y = np.zeros((len(sequences), vocab_size), dtype=np.bool_)

for i, seq in enumerate(sequences):
    for t, char in enumerate(seq):
        X[i, t, char2idx[char]] = 1
    y[i, char2idx[next_chars[i]]] = 1

# Define the model
model = Sequential([
    LSTM(128, input_shape=(seq_length, vocab_size)),
    Dense(vocab_size, activation='softmax')
])

model.compile(loss='categorical_crossentropy', optimizer='adam')
model.summary()

# Train the model
model.fit(X, y, batch_size=128, epochs=10)

# Sampling function with temperature
def sample(preds, temperature=1.0):
    preds = np.asarray(preds).astype('float64')
    preds = np.log(preds + 1e-8) / temperature
    exp_preds = np.exp(preds)
    preds = exp_preds / np.sum(exp_preds)
    return np.random.choice(range(len(preds)), p=preds)

# Generate text using trained model
def generate_text(seed_text, length=300, temperature=1.0):
    generated = seed_text
    for _ in range(length):
        input_seq = np.zeros((1, seq_length, vocab_size))
        for t, char in enumerate(seed_text):
            if char in char2idx:
                input_seq[0, t, char2idx[char]] = 1
        preds = model.predict(input_seq, verbose=0)[0]
        next_index = sample(preds, temperature)
        next_char = idx2char[next_index]
        generated += next_char
        seed_text = seed_text[1:] + next_char
    return generated

# Try generating text with different temperatures
seed = text[100:100+seq_length]
for temp in [0.2, 0.5, 1.0, 1.5]:
    print(f"\n--- Generating with temperature: {temp} ---\n")
    print(generate_text(seed, temperature=temp))
```

```python
#Question 4
import numpy as np
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load IMDB Dataset
max_features = 10000  # number of words to consider as features
maxlen = 200  # cut texts after this number of words (for padding)
embedding_dim = 128

(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=max_features)

# 2. Preprocess: Pad sequences
x_train = pad_sequences(x_train, maxlen=maxlen)
x_test = pad_sequences(x_test, maxlen=maxlen)

# 3. Define LSTM model
model = Sequential([
    Embedding(input_dim=max_features, output_dim=embedding_dim, input_length=maxlen),
    LSTM(128),
    Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()

# Train the model
model.fit(x_train, y_train, epochs=3, batch_size=64, validation_split=0.2)

# 4. Evaluate using confusion matrix and classification report
y_pred_probs = model.predict(x_test)
y_pred = (y_pred_probs > 0.5).astype("int32")

# Classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=["Negative", "Positive"]))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Negative", "Positive"], yticklabels=["Negative", "Positive"])
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()
```
